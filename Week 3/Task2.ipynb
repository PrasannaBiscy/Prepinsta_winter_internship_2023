```python
import pandas as pd
df=pd.read_csv("chipotle.csv",sep=",")
df
```

**1. Missing Values : Check for missing values in each column (Order ID, Quantity, Item Name, Choice Description, Item Price). How should missing values be handled?**

We can see that choice_description is the only column that has entries in it that need to have their missing data addressed. To handle the missing data, we'll fill the column with "description unavailable" for all entries having choice_description as Null.

```python
df.isnull().sum()
```

```python
df["choice_description"]=df["choice_description"].fillna("description unavailable")
df
```

**2. Data Types : Verify the data types of each column. Do they align with their expected types, and should any adjustments be made?**


```python
df.info()
```

**3. Duplicated Entries : Identify and handle duplicated entries in the dataset. How might duplicates impact analysis, and what is the appropriate action?**

I had to keep one instance of each entry after searching for entries with the same item name, which indicates multiple items, in order to handle duplicate entries.






```python
duplicate =df[df.duplicated(subset=['order_id','item_name'],keep=False)]
df=df.drop_duplicates(subset=['order_id','item_name'],keep='first')
df
```

**4. Quantity and Item Price : Examine the Quantity and Item Price columns. Are there any inconsistencies or anomalies that need correction?**

To find the number of products with the same price, I calculated the count of values whose quantity count and item_price are the same. Additionally, I computed the unique item_price values.




```python
print("\nQuantity:")
print(df['quantity'].value_counts())


print("\nItem Price:")
print(df['item_price'].value_counts())

print("\nItem Price:")
print(df['item_price'].unique())
```

**5. Choice Description : Analyze the Choice Description column. How should choices be handled, especially when there are multiple descriptions for a single item?**

To handle the numerous descriptions, we can concatenate them all into a single string that serves as the item's description.



```python
print("\nchoice_description:")
print(df['choice_description'].value_counts())
```

```python
df.loc[:, 'choice_description'] = df['choice_description'].str.replace(r'\[|\]', '')
df.loc[:, 'choice_description'] = df['choice_description'].str.split(', ').apply(', '.join)
df
```

**6. Handling Special Characters : Check for special characters in text-based columns (e.g., Item Name, Choice Description). How can these be addressed for consistency?**

'-' is a special character that appears in certain items in the item_name field. To be consistent, I've substituted " " for the '-'. For choice_descreption, solved it in Question No. 5.


```python
df.loc[:, 'item_name'] = df['item_name'].str.replace('-', ' ')
df
```

**7. Order Id Integrity : Cross-reference the Order ID column for integrity. Are there any irregularities or patterns that need validation?**

Certain items share the same order_id, but their item_name differs because we have already eliminated duplicate entries.



```python
order_id_integrity = df['order_id'].value_counts()
print("order ID's of single items : ")
single_item_id =order_id_integrity[order_id_integrity==1]
print(single_item_id)
print("\norder ID's of Multiple items : ")
multiple_item_id=order_id_integrity[order_id_integrity>1]
print(multiple_item_id)
```

**8. Item Name Standardization : Standardize the Item Name column. Are there variations that can be unified for better analysis?**

For uniformity, I removed the leading and trailing whitespaces.


```python
df.loc[:, 'item_name'] = df['item_name'].str.strip()
df
```

**9. Quantity and Price Relationships : Investigate the relationships between Quantity and Item Price. Are there cases where adjustments need to be made for accurate analysis?**

I have utilized group by to learn about the connections between various quantities and the price of the unique products that go with them.



```python
quantity_price_relationships = df.groupby(['quantity'])['item_price'].unique()

quantity_price_relationships
```

**10. Data Integrity Check : Perform a data integrity check by ensuring that quantities and prices align with the corresponding items and descriptions.**

I have grouped the values of item name and choice description with quantity and item price to see if they correspond with quantity and item price for the corresponding item name and choice description.The results demonstrate that the values are consistent.








```python
data_integrity_check = df.groupby(['item_name', 'choice_description'])[['quantity', 'item_price']].nunique()
data_integrity_check
```

**13. Consistent Quantity and Price Units : Ensure consistency in units for Quantity and Item Price. Should any conversions or adjustments be made for uniform analysis?**

Initially, I checked to see if any data were missing for the uniform analysis in either of the two columns. For the uniform analysis, I also converted the item price column from its original object format to a numerical representation.


```python
# checking for any missing values
null_quantity = df['quantity'].isnull().any()
null_item_price = df['item_price'].isnull().any()

print(null_quantity)
print(null_item_price)

print("------------------")

df.info()

print("------------------")

# converting price into numeric format

# Remove '$' sign, strip whitespace, and convert 'item_price' to float using .loc
df.loc[:, 'item_price'] = df['item_price'].replace("[$]", "", regex=True).str.strip().astype(float)


df.info()
```
