#### 2) Data Cleaning

```python
featured_columns= ['Make','Model','Variant','Ex-Showroom_Price','Displacement','Cylinders','Valves_Per_Cylinder','Drivetrain','Emission_Norm','Fuel_Type','Fuel_Tank_Capacity','Power','Torque','Kerb_Weight','Ground_Clearance','Front_Brakes','Rear_Brakes','Front_Suspension','Rear_Suspension','Front_Tyre_&_Rim','Rear_Tyre_&_Rim','Power_Steering','Power_Windows','Seating_Capacity','Type','Wheelbase','Wheels_Size','Central_Locking','Number_of_Airbags','ABS_(Anti-lock_Braking_System)','EBD_(Electronic_Brake-force_Distribution)','ESP_(Electronic_Stability_Program)','Tyre_Pressure_Monitoring_System']
clean_data = pd.DataFrame(df[featured_columns])
clean_data
```

```python
clean_data['Car Name'] = clean_data['Make'] + '-' + clean_data['Model']
# Converting 'Ex-Showroom_Price' column to int
clean_data['Ex-Showroom_Price'] = clean_data['Ex-Showroom_Price'].str.replace(',','')
clean_data['Ex-Showroom_Price'] = clean_data['Ex-Showroom_Price'].str.replace('Rs. ','')
clean_data['Ex-Showroom_Price'] = clean_data['Ex-Showroom_Price'].astype(int)
# Converting 'Displacement' column to int
clean_data['Displacement'] = clean_data['Displacement'].str.replace('cc','')
clean_data['Displacement'].fillna(0, inplace=True)
clean_data['Displacement'] = clean_data['Displacement'].astype(int)
clean_data.loc[:,'Wheelbase'] = clean_data['Wheelbase'].str.replace(' mm','')
clean_data = pd.DataFrame(clean_data)
clean_data
```

#### 5) Histograms

```python
numeric_columns = ['Ex-Showroom_Price','Displacement','Cylinders','Valves_Per_Cylinder','Seating_Capacity','Number_of_Airbags']
for col in numeric_columns:
    plt.hist(clean_data[col], color = 'pink', edgecolor = 'black')
    
    plt.title(f'Distribution of {col}')
    plt.xlabel(f'{col}')
    plt.ylabel('Frequency')
    plt.show()
```

#### 6) Bar Charts

```python
top_10_car_names = clean_data['Car Name'].value_counts().head(10)
# Bar chart for the distribution of car makers (top 10)
plt.figure(figsize=(20,6))
plt.bar(top_10_car_names.index,top_10_car_names.values,color = 'orange')
plt.title('Top 10 Car Names Distribution')
plt.xlabel('Car Name')
plt.ylabel('Values')
plt.show()
```

#### 7) Box Plots

```python
# Create box plots for each numeric variable
plt.figure(figsize=(15, 6))
for column in numeric_columns:
    plt.figure(figsize=(15, 6))
    #plt.subplot(num_rows, num_cols, i)
    sns.boxplot(x=clean_data[column], palette='muted')
    plt.title(f'Box Plot of {column}')
    plt.xlabel(f'{column}')
    plt.ylabel('Values')

plt.show()
```

#### 8) Pie charts

```python
top_5_car_models = clean_data['Model'].value_counts().head(5)
explode = (0.1,0,0,0,0,)
plt.figure(figsize=(6,6))
# create a pie chart
plt.pie(top_5_car_models.values, explode=explode,labels = top_5_car_models.index,autopct = '%1.1f%%', startangle=140)
plt.title('Distribution of Top 5 Car Models')
plt.show()
```

#### 9) Count Plots

```python
categorical_columns = ['Drivetrain','Fuel_Type','Seating_Capacity','Type']
# Set up subplots based on the number of columns
fig, axes = plt.subplots(nrows=len(categorical_columns), ncols=1, figsize=(10, 5 * len(categorical_columns)))

# Create count plots for each selected column
for i, column in enumerate(categorical_columns):
    sns.countplot(x=clean_data[column], ax=axes[i], order = clean_data[column].value_counts().index)
    axes[i].set_title(f'Count plot of {column}')
    axes[i].set_xlabel(f'{column}')
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

# Bivariate Analysis

#### 1) Correlation Matrix

```python
corr_matrix = clean_data[numeric_columns].corr()
# Visualize correlation matrix using heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix,annot = True, cmap = 'viridis',fmt = '.2f', linewidth=0.5)
plt.title('Correlation matrix of numeric variables')
plt.show()
```

#### 2) Point Plots

```python
# Distribution of 'Type' and 'Seating Capacity' and their relationship using joint plots
sns.set(style="whitegrid") 
sns.jointplot(x=clean_data['Type'], y=clean_data['Seating_Capacity'], data=clean_data, kind="scatter")
plt.show()
```

#### 3) Box Plots with Hue

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=clean_data['Type'], y=clean_data['Cylinders'], data=clean_data, hue="Drivetrain", palette="Set3")
plt.title('Box Plot for Type with Hue')
plt.show()
```

#### 4) Categorical Plots


```python
top_7_makes = clean_data['Make'].value_counts().nlargest(7).index

# Filter the DataFrame for the top makes
filter_data = clean_data[clean_data['Make'].isin(top_7_makes)]

# Create a stacked bar plot
plt.figure(figsize=(12, 8))
sns.countplot(x='Make', hue="Fuel_Type", data=filter_data, palette='bright', order=top_7_makes)
plt.title('Stacked Bar Plot for Top 10 Makes and Type')
plt.show()
```

#### 5) Violin Plots

```python
#Visualize the distribution of numeric Seating capacity across Drivetrain using violin plot
plt.figure(figsize=(12, 8))
sns.violinplot(x="Type", y="Cylinders", data=clean_data, palette="colorblind")
plt.title('Violin Plot for Seating Capacity across Drivetrain')
plt.show()
```

#### 6) Stacked bar charts

```python
top_7_model = clean_data['Car Name'].value_counts().nlargest(7).index

# Filter the DataFrame for the top makes
filtered_data = clean_data[clean_data['Car Name'].isin(top_7_model)]

# Create a stacked bar plot
plt.figure(figsize=(12, 8))
sns.countplot(x='Car Name', hue="Fuel_Type", data=filtered_data, palette="muted", order=top_7_model)
plt.title('Stacked Bar Plot for Top 7 car names and fuel type')
plt.show()
```
